```python
# author Grace Tseng
# version December 18, 2022

import csv
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt

EXIT_SUCCESS_CODE = 0   # inform the Bash shell of the exit value

def main():
  #admit rate graph
  data = pd.read_csv('Admit_Rate.csv')


  yearCount = data['Year']
  applyMale = data['Male']
  applyFemale = data['Female']
  admitMale = data['Male Admit']
  admitFemale = data['Female Admit']
  male = []
  female = []
  maleAdmit = []
  femaleAdmit = []
  totalApplicants = [] #total male and female applicants
  totalAdmits = [] #total num of admitted students
  percent = []
  maleAdmitRate = []
  femaleAdmitRate = []
  
  with open("Admit_Rate.csv", newline="") as file:
    reader = csv.reader(file, delimiter=",", quotechar="|")
    for row in reader:
        #maleAdmit.append(int(row[1])/int(row[3]))
        #femaleAdmit.append(row[2]/row[4])
        male.append(row[1])
        female.append(row[2])
        maleAdmit.append(row[3])
        femaleAdmit.append(row[4])


  male.pop(0)
  female.pop(0)
  maleAdmit.pop(0)
  femaleAdmit.pop(0)


  x = 1
  for i in male:
    combined = int(i) + int(female[x-1])
    totalApplicants.append(combined)
    x = x + 1

  k = 1
  for i in maleAdmit:
    combined = int(i) + int(femaleAdmit[k-1])
    totalAdmits.append(combined)
    x = k + 1

  j = 1
  for i in totalAdmits:
    combined = i/totalApplicants[j-1]*100
    percent.append(combined)
    j+= 1

  maleIterator = 1
  for i in maleAdmit:
    combined = int(i)/int(male[maleIterator-1])*100
    maleIterator += 1
    maleAdmitRate.append(combined)

  
  femaleIterator = 1
  for i in femaleAdmit:
    combined = int(i)/int(female[femaleIterator-1])*100
    femaleIterator += 1
    femaleAdmitRate.append(combined)

  plt.plot(yearCount, maleAdmitRate, label = "Male Admit Rate")
  plt.plot(yearCount, femaleAdmitRate, label = "Female Admit Rate")
  plt.plot(yearCount, percent, label = "Total Admit Rate")

  plt.legend()
  plt.title("Northwestern University Admission Statistics from years 2000 to 2021")
  plt.ylabel("Admit Rates in Percentage")
  plt.xlabel("Years (2000 - 2021)")

  plt.show()

  #SAT graph
  data = pd.read_csv('Scores.csv')
  yearCount = data['Year']
  SAT = data['SAT ']

  plt.plot(yearCount, SAT, label = "SAT")

  plt.legend()
  plt.title("Northwestern University SAT Statistics from 2000 to 2021")
  plt.ylabel("SAT Scores")
  plt.xlabel("Years (2000 - 2021)")

  plt.show()


  #ACT graph
  data = pd.read_csv('Scores.csv')
  yearCount = data['Year']
  ACT = data['ACT']

  plt.plot(yearCount, ACT, label = "ACT")

  plt.legend()
  plt.title("Northwestern University ACT Statistics from 2000 to 2021")
  plt.ylabel("ACT Scores")
  plt.xlabel("Years (2000 - 2021)")

  plt.show()

  # exit mainline with success code, 0x00
  exit(EXIT_SUCCESS_CODE)


  # Python's mainline for your program
if __name__ == '__main__':
    main()

  
```
